CONVERTING DATETIME VALUES

```python
import pandas as pd
df=pd.read_csv('employees.csv',parse_dates=["Last Login Time"])
#parse_dates -> d/m/y
#to_datetime -> y/m/d
pd.to_datetime(df["Start Date"])
```

FILTERING A DF WITH MORE THAN ONE CONDITION with AND(&)

```python
mask3=df["Team"]=="Legal"
mask4=df["Salary"]>100000
```

```python
df[mask3&mask4]
```

OR CONDITION: |

```python
df[mask3|mask4]
```

.DUPLICATED() METHOD 

```python
import pandas as pd
df=pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"])
```

```python
mask=~df["First Name"].duplicated(keep=False) #this will remove all False ( to fulfill False, it has to be duplicates) and leave 
                                              #only unique values
df[mask]
```

drop_duplicates() method

```python
df.drop_duplicates(subset=["Team"],keep="first")
df.drop_duplicates(subset=["First Name"],keep=False)
```

```python
mask6=~df.duplicated("First Name",keep=False)
```

```python
df[mask6]
```
